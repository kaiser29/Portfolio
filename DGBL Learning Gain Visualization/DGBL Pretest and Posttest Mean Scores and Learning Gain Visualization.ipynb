```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statistics import mean
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('./DGBL Learning Gain DataSet.csv')
df.head()
```

```python
df.drop('Mean difference', inplace=True, axis =1)
df.drop('4.27027027', inplace=True, axis =1)
df = df
df.head()
```

# Mean Difference of Pretest and Posttest (Easy, Average, Difficult)

```python
pre_easy = df['Pretest(Easy)'].mean()
```

```python
post_easy = df['Posttest(Easy)'].mean()
post_easy
```

```python
pre_ave = df['Pretest(Average)'].mean()
post_ave = df['Posttest(Average)'].mean()

print(pre_ave)
print(post_ave)
```

```python
pre_diff = df['Pretest(Difficult)'].mean()
post_diff = df['Posttest(Difficult)'].mean()

print(pre_diff)
print(post_diff)
```

```python
df.describe()
```

```python
N = 3
ind = np.arange(N) 
width = 0.25


#Pretest Mean(Easy, Average, Difficult)
xvals = [7.864865,5.459459,5.540541]
bar1 = plt.bar(ind, xvals, width, color = 'r')

#Posttest Mean(Easy, Average, Difficult)
yvals = [8.513514,7.459459, 7.162162]
bar2 = plt.bar(ind+width, yvals, width, color='g')

plt.xticks(range(len(xvals)), ('5(Easy)','5(Average)','5(Difficult)'))

  
plt.xlabel('Scores')
plt.ylabel('Scores')
plt.title("Mean Scores of Pretest, Posttest")
  

plt.legend( (bar1, bar2), ('Pretest Mean Score', 'Posttest Mean Score') )
plt.show()
```
