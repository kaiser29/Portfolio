```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
import seaborn as sns

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
movies = pd.read_csv('./movies.csv')
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
newmovies = movies.dropna(axis = 0)
```

```python
newmovies.head()
```

```python
newmovies.isnull().sum()
```

```python
newmovies['votes'] = newmovies['votes'].astype('int64')
newmovies['budget'] = newmovies['budget'].astype('int64')
newmovies['gross'] = newmovies['gross'].astype('int64')
newmovies['runtime'] = newmovies['runtime'].astype('int64')
```

```python
newmovies.head()
```

```python
sns.regplot(x="budget", y="gross", data=newmovies, scatter_kws={"color": "purple"}, line_kws={"color":"black"})
```

```python
corr_matrix = newmovies.corr(method = 'pearson')

corr_matrix
```

```python
sns.heatmap(corr_matrix, annot=True)

plt.title("Correlation Matrix for Numerical Values")

plt.xlabel("Movie Features")

plt.ylabel("Movie Features")

plt.show()
```

```python
corr_pairs = corr_matrix.unstack()

sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
